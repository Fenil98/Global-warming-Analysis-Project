# EDA for CO2 emissions (metric tons per capita) dataset

The dataset is collected from **[CO2 emissions](https://data.worldbank.org/indicator/EN.ATM.CO2E.PC)**. 
        
    Carbon dioxide emissions are those stemming from the burning of fossil fuels and the manufacture of cement. 
    They include carbon dioxide produced during the consumption of solid, liquid, and gas fuels and gas flaring.

## 6. To check outliers in the dataset

```python
sns.boxplot(data = co2)
plt.title(' Outliers', color = 'blue', size = 18)
plt.xlabel('Columns', color = 'red')
plt.ylabel('Count', color = 'red')
```

## 7. CO2 emission in year 1965 for sample dataset of 5 country

```python
sample = co2.sample(n = 5)

sample
```

```python
year = sample['1965']

cc = sample['Country Code']
```

```python
plt.bar(cc, year)
plt.title(' CO2 emissions (metric tons per capita) for year 1965', color = 'blue', size = 18)
plt.xlabel('Country Code', color = 'red', size = 12)
plt.ylabel('mt per capita', color = 'red', size = 12)
```
