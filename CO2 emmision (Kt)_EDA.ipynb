# EDA for CO2 emissions (kt) dataset

You can find this dataset **[CO2 emissions](https://data.worldbank.org/indicator/EN.ATM.CO2E.KT)** here. 

    Carbon dioxide emissions are those stemming from the burning of fossil fuels and the manufacture of cement. 
    They include carbon dioxide produced during the consumption of solid, liquid, and gas fuels and gas flaring.

## 6. To check outliers in the dataset

```python
sns.boxplot(data = Co2_in_kt)
plt.title(' Outliers', color = 'blue', size = 18)
plt.xlabel('Columns', color = 'red')
plt.ylabel('Count', color = 'red')
```

## 7. CO2 emissions (kt) in year 2000 for sample dataset of 5 country

```python
sample = Co2_in_kt.sample(n = 5)

sample
```

```python
year = sample['2000']

cc = sample['Country Code']
```

```python
plt.bar(cc, year)
plt.title('CO2 emissions (kt) in year 2000', color = 'blue', size = 18)
plt.xlabel('Country code', color = 'red', size = 12)
plt.ylabel('Co2 emmission in KT', color = 'red', size = 12)
```
