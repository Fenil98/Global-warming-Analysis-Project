# Changelog
This file contains the notable changes to the project for all 5 datasets. 

## GDP dataset

### Version 1.0.0 (07-26-2021)

#### New
Performed transpose and added one column "year"

#### Changes
Removed this two columns "Indicator Name", "Indicator Code"

Changes in columns name { 'Country Name': 'country', 'level_1': 'year', 0:'current GDP' }

changed column type {"country" : str, "year" : int}
