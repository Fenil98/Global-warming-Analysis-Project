# EDA for GDP dataset

If you are looking for a dataset than it is [here!](https://data.worldbank.org/indicator/NY.GDP.MKTP.CD)
    
    GDP at purchaser's prices is the sum of gross value added by all resident producers in the economy plus any product taxes. Data are in current U.S. dollars. Dollar figures for GDP are converted from domestic currencies using single-year official exchange rates.

## 6. To check outliers in the dataset

```python
sns.boxplot(data = GDP)
plt.title(' Outliers', color = 'blue', size = 18)
plt.xlabel('Columns', color = 'red')
plt.ylabel('Count', color = 'red')
```

## 7. GDP in year 2011 for sample dataset of 5 country

```python
sample = GDP.sample(n = 5)

sample
```

```python
year = sample['2011']

cc = sample['Country Code']
```

```python
plt.bar(cc, year)
plt.title('GDP in year 2011', color = 'blue', size = 18)
plt.xlabel('Country Code ', color = 'red', size = 12)
plt.ylabel('US$', color = 'red', size = 12)
```
