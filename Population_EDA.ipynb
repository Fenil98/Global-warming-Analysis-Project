# EDA for Population dataset

[This](https://data.worldbank.org/indicator/SP.POP.TOTL?end=2019&name_desc=false&start=1960) is the source of our population dataset.

    Total population is based on the de facto definition of population, which counts all residents regardless of legal status or citizenship. The values shown are midyear estimates.
   Visit [de facto definition](https://ec.europa.eu/home-affairs/what-we-do/networks/european_migration_network/glossary_search/de-facto-population_en) to know more about de facto population.

## 6. To check outliers in the dataset

```python
sns.boxplot(data = pop)
plt.title(' Outliers', color = 'blue', size = 18)
plt.xlabel('Columns', color = 'red')
plt.ylabel('Count', color = 'red')
```

## 7. Population in year 2002 for sample dataset of 5 country

```python
sample = pop.sample(n = 5)

sample
```

```python
year = sample['2002']

cc = sample['Country Code']
```

```python
plt.bar(cc, year)
plt.title('Population in year 2002', color = 'blue', size = 18)
plt.xlabel('Country Code', color = 'red', size = 12)
plt.ylabel('population in million', color = 'red', size = 12)
```
